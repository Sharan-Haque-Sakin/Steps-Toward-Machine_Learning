# Linear Regression with Single Variable/Feature

```python
df = pd.read_csv("/kaggle/input/area-home-price/homeprices.csv")
df.head()
```

```python
rg = linear_model.LinearRegression()
rg.fit(df[["area"]] , df.price)
```

```python
rg.predict(pd.DataFrame({ "area": [3200]}))
```

```python
plt.scatter(df.area, df.price , color="red" , marker="+")
plt.xlabel("area (sq ft)")
plt.ylabel("Prices (Dollar $)")
plt.title(" Area vs Price Graph")
plt.plot(df.area , rg.predict(df[["area"]]) , color="blue")
```

# Linear Regression with Multiple Variables/Features

```python
DF = pd.read_csv("/kaggle/input/housedata/data.csv")
DF.head(6)
```

```python
RG = linear_model.LinearRegression()

RG.fit(DF[["bedrooms","bathrooms","sqft_living","floors"]] , DF.price)
```

```python


pred_df = pd.DataFrame({
    "bedrooms": [2],
    "bathrooms": [1],
    "sqft_living": [2000],
    "floors": [2]
})

RG.predict(pred_df)
```
